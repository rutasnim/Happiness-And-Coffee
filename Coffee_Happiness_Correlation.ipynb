# Coffee and Happiness Correlation

### Data& Libraries import

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Coffee data loading
coffee= pd.read_csv(r'Desktop/Projects/Coffee/Coffee.csv', sep=",")
```

```python
coffee.head()
```

```python
#Loading Happiness data 

score15=pd.read_csv(r'Desktop/Projects/Coffee/2015.csv', sep=",")
score16=pd.read_csv(r'Desktop/Projects/Coffee/2016.csv', sep=",")
score17=pd.read_csv(r'Desktop/Projects/Coffee/2017.csv', sep=",")
score18=pd.read_csv(r'Desktop/Projects/Coffee/2018.csv', sep=",")
score19=pd.read_csv(r'Desktop/Projects/Coffee/2019.csv', sep=",")
```

### Data Cleaning


```python
#Coffee dropping all the NaN 
coffee.dropna()
```

```python
#joining happiness score 
frames= [score15, score16,score17, score18, score19]
scores=pd.concat(frames)
```

```python
scores.head()
```

## Join Coffee and Happiness dataframes

```python

df=pd.merge(coffee, scores, left_on=["Country","Year"], right_on=["Country","Year"], how="inner")
```

```python
df.count()
```

## Correlation

```python
df.corr()>0.5
```

```python
r=df["Happiness Score"].corr(df['Cups_consumed_per_capita'])
r
```

```python
sns.set(rc={'figure.figsize':(10,6)})
```

```python
sns.regplot(x="Happiness Score",y='Cups_consumed_per_capita',  data=df)
plt.title("Coffee Consumption vs. Happines score of Countries in 2015-2019", size=15)
plt.text(5.5,2470,'Correlation Coeff,r=0.65', fontsize=15, color='r', visible=True, bbox=dict(boxstyle="square", color='w'))
plt.savefig('corr.png', bbox_inches='tight')
```
