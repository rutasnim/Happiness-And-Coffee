### In this notebook, I am going to do the following:

- Change the country name with special characters since Mysql that I had installed couldn't handle them 
- Change the country name to match with country names of the coffee consumption table in order to perform join and analysis.
- Change column names so MYsql can handle them


```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv(r'Desktop/Projects/Coffee/Population.csv', sep=",")
```

```python
df
```

## Looking at all country names:

```python
df['Country'].unique()
```

Country names that needs to be changed
- DR Congo -> Congo

- Côte d'Ivoire -> Ivory Coast
- Czech Republic( Czechia) -> Czech Republic
- Curaçao
- Timor-Leste-> East Timor
- Réunion-> Reunion

```python
df.loc[df['Country']=='DR Congo', 'Country'] ='Congo'

df.loc[df['Country']=="Côte d'Ivoire", 'Country'] ='Ivory Coast'
df.loc[df['Country']=='Czech Republic (Czechia)', 'Country'] ='Czech Republic'
df.loc[df['Country']=='Curaçao', 'Country'] ='Curacao'
df.loc[df['Country']=='Timor-Leste', 'Country'] ='East Timor'

df.loc[df['Country']=='Réunion', 'Country'] ='Reunion'
```

```python
df['Country'].unique()
```

## Renaming coloumns

```python
df.rename(columns={'Yearly % Change': 'yearly_change_percent', 'Yearly Change': 'yearly_change', 'Migrants (net)':'net_migrant','Median Age':'median_age','Fertility Rate': 'fertility_rate','Density (P/Km²)':'density','Urban Pop %':'urban_pop_percent','Urban Population':'urban_population',"Country's Share of World Pop %":'country_share_of_the_world_pop_percent','World Population':'world_population','Country Global Rank':'country_global_rank'}, inplace=True)
```
